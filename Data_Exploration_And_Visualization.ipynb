# Lets categorize the majors. First we begin with STEM majors. Read more about STEM majors here

https://en.wikipedia.org/wiki/Science,_technology,_engineering,_and_mathematics

```python
#stem_cats==stem category
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
```

```python
fig=plt.figure(figsize=(20,3))
for sp in range(0,6):
    ax=fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees["Year"],women_degrees[stem_cats[sp]],c=cb_dark_blue,label='women',
    linewidth=3)
    ax.plot(women_degrees["Year"],100-women_degrees[stem_cats[sp]],c=cb_orange,label="men",
    linewidth=3)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off",top="off",right="off",left="off")
    if sp==0:
        ax.text(2005,87,"Men")
        ax.text(2002,8,"Women")
    elif sp==5:
        ax.text(2005,62,"Men")
        ax.text(2001,35,"Women")
plt.show()        
```

From the graphs, it is clear that the gender gap between men and women is smallest among Math and Stats majors

# Seaborn

This is another powerfull vizualization libaray built on top of matplotlib. Details about this libaray can be found here http://seaborn.pydata.org/index.html

```python
import seaborn as sns
```

Seaborn has different plotting styles. However, it is important to choose one that doesn't distract the eye from observing with ease what the story the data conveys. More info about sytles can be found here https://seaborn.pydata.org/tutorial/aesthetics.html#styling-figures-with-axes-style-and-set-style 

```python

sns.set_style("white")
```

Lets set the font size


```python
sns.set(font_scale=1.5) 
```

Boxplots are often used to vizualize the distribution of data. It comes very handy when visualizing categorical data. Histograms could be used too. However, quantities such as lower quartile etc are not immediately obvious from a histogram. When comparing graphs, have the limits on the axis of all graphs equal aids visual comparison. 

```python

fig = plt.figure( figsize=(20,18))
fig.subplots_adjust(hspace=0.1)
ax1 = fig.add_subplot(2,3,1)
ax2 = fig.add_subplot(2,3,2)
ax3=fig.add_subplot(2,3,3)
ax4=fig.add_subplot(2,3,4)
ax5=fig.add_subplot(2,3,5)
ax6=fig.add_subplot(2,3,6)

ax1=sns.boxplot(data=women_degrees[stem_cats],ax=ax1,saturation=0.4,color="b",linewidth=2)
ax1.set_xticklabels([])
ax1.set_ylim(0,100)
ax1.set_yticks([0,100])
ax1.set_title("Stem Category")


ax2=sns.boxplot(data=women_degrees[lib_arts_cats],ax=ax2,saturation=0.4,color="b",linewidth=2)
ax2.set_xticklabels([])
ax2.set_ylim(0,100)
ax2.set_yticks([0,100])
ax2.set_title("Liberal Arts Category")

ax3=sns.boxplot(data=women_degrees[other_cats],ax=ax3,saturation=0.4,color="b",linewidth=2)
ax3.set_xticklabels([])
ax3.set_ylim(0,100)
ax3.set_yticks([0,100])
ax3.set_title("Other Category")


ax4=sns.boxplot(data=100-women_degrees[stem_cats],ax=ax4,saturation=0.4,color="orange",linewidth=2)
ax4.set_xticklabels(stem_cats,rotation=90)
ax4.set_ylim(0,100)
ax4.set_yticks([0,100])

ax5=sns.boxplot(data=100-women_degrees[lib_arts_cats],ax=ax5,saturation=0.4,color="orange",linewidth=2)
ax5.set_xticklabels(lib_arts_cats,rotation=90)
ax5.set_ylim(0,100)
ax5.set_yticks([0,100])

ax6=sns.boxplot(data=100-women_degrees[lib_arts_cats],ax=ax6,saturation=0.4,color="orange",linewidth=2)
ax6.set_xticklabels(other_cats,rotation=90)
ax6.set_ylim(0,100)
ax6.set_yticks([0,100])


sns.despine(top=True, bottom=True, left=True, right=True)
```

The top row is the distribution for women while the bottom row is for men. From the figure, the distribution for physical sciences seem equally spread. Lets generate a plot showing the line graph and disbribution side by side 


```python
fig=plt.figure(figsize=(15,7))
ax = fig.add_subplot(1,2,1)
ax.plot(women_degrees["Year"],women_degrees["Physical Sciences"],
           c=cb_dark_blue,label="women",linewidth=3)
ax.plot(women_degrees["Year"],100-women_degrees["Physical Sciences"],
           c=cb_orange,label="men",linewidth=3)
for key, spine in ax.spines.items():
        spine.set_visible(False)
ax.set_xlim(1968,2011)
ax.set_ylim(0,100)
ax.set_yticks([0,100])
ax.set_yticklabels([0,100])
ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
ax.tick_params(bottom="off",top="off",left="off",right="off")
ax.tick_params(labelbottom="on")
plt.legend(loc="upper left")

ax2=fig.add_subplot(1,2,2)
ax2=sns.boxplot(data=women_degrees["Physical Sciences"],ax=ax2,saturation=0.4,color="b",linewidth=2)
ax2=sns.boxplot(data=100-women_degrees["Physical Sciences"],ax=ax2,saturation=0.6,color="orange",linewidth=2)
ax2.set_xticklabels([])
ax2.set_ylim(0,100)
ax2.set_yticks([0,100])
plt.suptitle('Line graph and box plot of Physical Sciences')
```
